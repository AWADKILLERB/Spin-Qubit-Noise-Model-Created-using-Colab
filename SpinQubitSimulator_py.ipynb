```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_qubit_noise(detuning_noise_amplitude, time_steps=1000):
    """
    محاكاة بسيطة لتأثير الضجيج على استقرار الكيوبت (Qubit Fidelity)
    بناءً على نموذج Hamiltonian الخاص بك.
    """
    t = np.linspace(0, 10, time_steps)
    # نموذج مبسط لتطور الحالة تحت تأثير الضجيج
    noise = np.random.normal(0, detuning_noise_amplitude, time_steps)
    fidelity = np.exp(-np.cumsum(noise**2) / time_steps)

    return t, fidelity

# محاكاة لثلاث مستويات مختلفة من الضجيج
amplitudes = [0.01, 0.05, 0.1]
plt.figure(figsize=(8, 5))

for amp in amplitudes:
    t, f = simulate_qubit_noise(amp)
    plt.plot(t, f, label=f'Noise Amplitude: {amp}')

plt.title('EDSR Qubit Fidelity under Detuning Noise')
plt.xlabel('Time (a.u.)')
plt.ylabel('Average Gate Fidelity')
plt.legend()
plt.grid(True)
plt.savefig('fidelity_plot.png') # هذا الملف سيظهر في الـ README
plt.show()
```
